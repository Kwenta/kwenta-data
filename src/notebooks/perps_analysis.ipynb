```python
import pandas as pd
import sqlite3
from datetime import datetime
import plotly.express as px
```

```python
# Read sqlite query results into a pandas DataFrame
con = sqlite3.connect("../../data/perps.db")
```

```python
df_trades = pd.read_sql_query("SELECT * from trades", con)
df_transfers = pd.read_sql_query("SELECT * from transfers", con)
df_positions = pd.read_sql_query("SELECT * from positions", con)
df_stats = pd.read_sql_query("SELECT * from stats", con)
df_funding = pd.read_sql_query("SELECT * from funding_rates", con)
```

```python
df_trades['trackingCode'] = df_trades['trackingCode'].apply(lambda x: bytearray.fromhex(x[2:]).decode().replace('\x00', ''))
df_trades['day'] = df_trades['timestamp'].apply(lambda ts: datetime.utcfromtimestamp(int(ts)).strftime('%Y-%m-%d'))
df_trades['volume'] = abs(df_trades['price'] * df_trades['size'])
df_trades['trackingCode'] = df_trades['trackingCode'].replace('', 'No tracking code')

df_trades
```

```python
df_tracking = df_trades.groupby(['day', 'trackingCode'])['volume'].sum().reset_index()

fig = px.bar(
    df_tracking,
    x='day',
    y='volume',
    color='trackingCode'
)
fig.show()
```
